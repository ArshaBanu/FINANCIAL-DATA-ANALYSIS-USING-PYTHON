## INSIGHTS GENERATION AND REPORT
This section summarises the key insights derived from the exploratory data analysis (EDA) conducted on Bank of America’s consumer complaints from 2017 to 2023. The objective is to translate the analytical findings into clear, meaningful insights that can be understood by both technical and non-technical readers. This phase highlights major patterns, trends, and anomalies discovered in the data and concludes with recommendations and next steps based on the results.


###  How the Insights Were Developed?
To generate insights, the dataset was examined using the following approaches:
- Descriptive statistics to quantify the most common products, issues, submission methods, and response types.
- Group-level comparisons (e.g., complaints by product, by issue, by year, by month).
- Visual exploration using bar charts, line charts, pie charts, and heatmaps.
- Multidimensional analysis (product–issue and product–timeliness relationships).
- Time-based trend analysis to understand how complaint patterns evolve across years and months.


Because the dataset contains only one numerical field (Complaint ID), which is an identifier rather than a meaningful metric, statistical correlation was not included. Instead, relationships between categorical variables were explored to reveal operational patterns.


## Key Insights

### 1. Most Complaints Come from Core Banking Products
The analysis showed that the majority of complaints come from two major product categories—checking or savings accounts and credit card or prepaid card products. These products together account for a large portion of all complaints submitted between 2017 and 2023. This is expected because these are essential, high-frequency banking services used daily by millions of customers. Any delay, technical issue, unexpected fee, or service disruption in these products has an immediate impact on customers, leading to more complaints compared with more specialized products like student loans or vehicle loans. This insight highlights the importance of strengthening everyday banking services, as improvements here will directly improve customer experience for the largest user groups.



### 2. Managing an Account Is the Most Common Issue
Across the entire dataset, “Managing an account” stands out as the single most common issue raised by customers. This issue appears far more frequently than any other complaint type, indicating that customers regularly face challenges related to account access, functionality, usability, or accuracy. The problem may involve difficulties using online banking tools, confusion about account changes, struggles with account security, or delays in resolving account-related actions. Notably, this issue increased sharply in 2023, suggesting new or worsening problems in how customers interact with their accounts. This insight points toward a systemic issue in account management processes or digital interfaces, making it a crucial area for Bank of America to investigate and address..




### 3. Complaints Show a Clear Seasonal Trend, Peaking in July Each Year
Complaint activity is not constant throughout the year. Instead, the data shows a consistent seasonal pattern, with complaints rising through the first half of the year and peaking in July. After this peak, the number of complaints decreases sharply in August and particularly in September. This pattern appears repeatedly across multiple years, suggesting predictable annual pressure points. The mid-year surge might be due to increased customer activity, system upgrades, staff turnover cycles, policy changes, or seasonal financial events such as holiday bill repayments or mid-year travel spending. Understanding this seasonal behaviour can help the bank plan staffing, support coverage, and system resilience during peak complaint periods.




### 4. Complaint Volumes Increased Until 2022, Then Declined in 2023
Looking at complaint trends over time, there is a clear pattern of steadily rising complaints from 2017 up to 2022, which recorded the highest number of complaints across the entire dataset. However, 2023 shows a noticeable drop in complaint volume, though still higher than the earliest years in the dataset. The consistent growth until 2022 could be linked to higher digital engagement, increased customer awareness of complaint channels, or actual worsening of service quality in certain areas. The decline in 2023 may reflect improved service delivery, corrective actions taken by the bank, or external factors that reduced complaint activity. This insight highlights a shift in customer behaviour and operational performance over time.




### 5. Customers Strongly Prefer Submitting Complaints Online
The analysis of submission methods revealed that over 70% of complaints are submitted through the Web, making it by far the most preferred channel. Other channels, such as phone, referrals, postal mail, email, and fax, account for much smaller proportions. This indicates a strong reliance on online complaint services and reflects modern customer expectations for fast, convenient, and accessible digital solutions. The dominance of online submission suggests that any weaknesses in the digital complaint portal—such as slow response times, technical issues, or unclear instructions—could significantly affect customer satisfaction. It also shows that maintaining a reliable online system is essential for providing efficient customer service.




### 6. Most Complaints Are Closed with Explanation (Low Monetary Relief)
A large proportion of complaints are resolved using the response category “Closed with explanation,” meaning the bank addresses the issue by providing clarification rather than offering monetary or non-monetary relief. This pattern suggests that many complaints may stem from misunderstandings, confusion about policies, or lack of clarity in communication rather than actual financial harm or service errors. While this can be positive—indicating that many issues can be resolved through communication—it also raises questions about whether explanations alone satisfy customers or if additional follow-up, clearer documentation, or better customer education is needed to reduce repeated complaints on the same topics.




### 7. Responses Are Mostly Timely, but 2021 Shows a Significant Spike in Delays
Overall, Bank of America performs well in terms of response time, with the majority of complaints receiving a timely response. However, the analysis found a significant spike in untimely responses in 2021, making that year an outlier. This unusual increase could indicate internal challenges, such as operational disruptions, system changes, staffing shortages, or heightened customer demand. The improvement in 2022 and 2023 suggests that the issue may have been addressed, but the spike remains an important historical indicator of system or process stress. Monitoring future timeliness trends will be important to ensure similar disruptions do not recur.




### 8. Product–Issue Heatmaps Show Clear Areas of Concentration
The heatmap analysis revealed that certain product-issue combinations are highly concentrated, meaning they occur far more frequently than others. For example, “Managing an account” is strongly associated with checking/savings accounts and credit cards. Credit reporting problems cluster heavily under credit-reporting services, while transaction-dispute issues are more common under credit card products. These clusters reveal exactly where the bank should focus its improvement efforts, as they highlight the intersection between specific products and the issues most likely to affect customers. Such insight is valuable for prioritising resources, designing training programs, and implementing targeted process improvements.



## Summary of Significant Patterns and Anomalies
Across the dataset, several patterns are clear:
- A small number of products account for most complaints.
- A single issue type dominates complaints across multiple years.
- A repeating seasonal spike in July indicates recurring mid-year stress.
- A major timeliness anomaly in 2021 highlights a period of operational strain.
- Explanatory responses dominate, suggesting non-financial resolutions are the norm.


These patterns point toward systemic challenges in account management and customer communication.


## Overall Findings
In summary, Bank of America’s complaint trends from 2017 to 2023 show that customers face persistent issues with managing their accounts, particularly within core banking products. Complaint volumes increased steadily until 2022 and follow a consistent seasonal cycle, peaking mid-year. Most customers prefer to submit complaints online, and most complaints are handled on time. However, the sharp increase in untimely responses in 2021 and the overwhelming use of “Closed with explanation” responses signal areas that require closer review. Overall, the findings highlight both strengths in responsiveness and weaknesses in account-related customer experiences.


## Recommendations and Next Steps
### 1. Investigate Account-Management Complaints in Detail
Because “Managing an account” is by far the most common issue—and has increased sharply in recent years—this area requires immediate and thorough investigation. This may involve reviewing the entire customer journey, including how customers log in, navigate their accounts, view balances, initiate transfers, or update personal information. The bank may need to analyse whether there are recurring system errors, confusing interface designs, or inconsistent communication that lead customers to feel uncertain or frustrated. A detailed review of digital banking tools, mobile app usability, and customer support documentation could help uncover the root causes of these problems. Improving this area will likely reduce a large portion of the bank’s overall complaints.


### 2. Prioritise Improvements for High-Volume Products
Since checking/savings accounts and credit cards generate the greatest number of complaints, improvements in these products will have the highest impact on customer satisfaction. The bank should consider prioritising enhancements such as clearer fee explanations, smoother transaction processing, more intuitive app features, and faster resolution workflows for these products. Additionally, specialised training for customer service teams handling these categories may help resolve issues more effectively. By focusing resources on these high-volume products, the bank can address the most widespread sources of frustration and demonstrate a strong commitment to customer care.


### 3. Analyse the Causes Behind the 2021 Timeliness Spike
The unexpected rise in untimely responses in 2021 suggests that something disrupted normal operations during that period. Understanding what happened—and why—can help the bank prevent similar issues in the future. This analysis may involve reviewing internal records regarding staffing levels, customer service workloads, technology upgrades, policy changes, or external events (such as regulatory shifts or economic pressures). If the spike was caused by system outages or resource shortages, the bank should consider strengthening contingency plans, increasing automation, or improving internal coordination. Learning from this anomaly can help build a more resilient complaint-handling process.


### 4. Re-evaluate “Closed with Explanation” as a Primary Resolution
Although “Closed with explanation” is the most common response type, it is important to determine whether explanations genuinely solve customers’ concerns or simply close the complaint. Customers may still feel dissatisfied if explanations lack clarity, fail to address the root issue, or do not offer next steps. The bank could consider analysing customer feedback, repeat complaints, or escalation rates to assess the effectiveness of explanations. Clearer communication templates, proactive follow-ups, or supplementary resources (e.g., FAQs, help guides) may be needed. Strengthening this area can help reduce repeated frustration and improve customer trust.


### 5. Monitor Complaint Trends Continuously
To maintain consistent oversight of complaint trends, the bank would benefit from creating dashboards or automated monthly/quarterly reports. These tools can display real-time data on complaint volumes, response times, issue categories, seasonal patterns, and emerging spikes. With this type of continuous monitoring, managers can quickly identify developing issues before they grow into larger problems. For example, if account-management complaints surge suddenly, the bank can intervene immediately rather than waiting until the problem becomes widespread. A systematic monitoring approach supports faster decision-making, more efficient operations, and long-term service improvement.
